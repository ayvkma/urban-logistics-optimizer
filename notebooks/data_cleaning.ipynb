```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/tokyo_orders_raw.csv')
initial_row_count = len(df)
```

```python
df.head()
```

```python
initial_dups = df.duplicated().sum()
df.drop_duplicates(inplace=True)
removed_dups = initial_dups - df.duplicated().sum()
```

```python
nulls_cnt = df[['latitude', 'longitude']].isnull().sum().sum()
df = df.dropna(subset=['latitude', 'longitude'])
```

```python
MIN_LATITUDE = 35.5000
MAX_LATITUDE = 35.8000
MIN_LONGITUDE = 139.5000
MAX_LONGITUDE = 139.9000
outliers_cnt = df[~(df['latitude'].between(MIN_LATITUDE, MAX_LATITUDE) & df['longitude'].between(MIN_LONGITUDE, MAX_LONGITUDE))].shape[0]
df = df[df['latitude'].between(MIN_LATITUDE, MAX_LATITUDE) & df['longitude'].between(MIN_LONGITUDE, MAX_LONGITUDE)]
```

```python
print(f"Initial rows: {initial_row_count}")
print(f"Outliers removed: {outliers_cnt}")
print(f"Null values removed: {nulls_cnt}")
print(f"Duplicate rows removed: {removed_dups}")
print(f"Final rows: {len(df)}")
```

```python
df.to_csv('../data/processed/tokyo_orders_clean.csv', index=False)
```
